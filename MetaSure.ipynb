```python
import os
import cv2

# Define paths
original_videos_folder = "/content/DFD_original_sequences"
output_frames_folder = "/content/extracted_frames_original"

# Create output folder if it doesn't exist
os.makedirs(output_frames_folder, exist_ok=True)

# Set parameters
frame_interval = 10  # Extract 1 frame every 10 frames
max_frames = 3000
frame_count = 0

# Loop through videos
for video_file in os.listdir(original_videos_folder):
    if frame_count >= max_frames:
        break

    video_path = os.path.join(original_videos_folder, video_file)
    cap = cv2.VideoCapture(video_path)

    frame_idx = 0
    while cap.isOpened() and frame_count < max_frames:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_idx % frame_interval == 0:
            frame_name = f"{video_file}_frame_{frame_idx}.jpg"
            cv2.imwrite(os.path.join(output_frames_folder, frame_name), frame)
            frame_count += 1
        frame_idx += 1

    cap.release()

print(f"Extraction complete! {frame_count} frames saved in {output_frames_folder}.")
```

```python
import cv2
import os

# Define input and output folders
video_folder = "/content/DFD_manipulated_sequences"
output_folder = "/content/manipulated_images"

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Set frame extraction limit
frame_limit = 3000
frame_count = 0

# List all video files
video_files = [f for f in os.listdir(video_folder) if f.endswith(('.mp4', '.avi', '.mov'))]

for video_file in video_files:
    if frame_count >= frame_limit:
        break  # Stop when we reach the limit

    video_path = os.path.join(video_folder, video_file)
    cap = cv2.VideoCapture(video_path)

    success, frame = cap.read()
    frame_id = 0

    while success and frame_count < frame_limit:
        frame_filename = os.path.join(output_folder, f"{video_file}_frame_{frame_id:04d}.jpg")
        cv2.imwrite(frame_filename, frame)
        frame_count += 1
        frame_id += 1

        success, frame = cap.read()

    cap.release()

print(f"Extraction complete. {frame_count} frames saved in {output_folder}")
```

```python
import os
import shutil
import random

# Define paths
dataset_path = "/content/dataset"
train_path = "/content/dataset/train"
val_path = "/content/dataset/val"

# Classes
classes = ["original", "manipulated"]

# Create train/val directories
for cls in classes:
    os.makedirs(os.path.join(train_path, cls), exist_ok=True)
    os.makedirs(os.path.join(val_path, cls), exist_ok=True)

# Function to split images
def split_data(source_dir, train_dir, val_dir, split_ratio=0.8):
    all_files = os.listdir(source_dir)
    random.shuffle(all_files)

    split_index = int(len(all_files) * split_ratio)
    train_files = all_files[:split_index]
    val_files = all_files[split_index:]

    for file in train_files:
        shutil.move(os.path.join(source_dir, file), os.path.join(train_dir, file))

    for file in val_files:
        shutil.move(os.path.join(source_dir, file), os.path.join(val_dir, file))

# Perform the split
for cls in classes:
    split_data(
        source_dir=os.path.join(dataset_path, cls),
        train_dir=os.path.join(train_path, cls),
        val_dir=os.path.join(val_path, cls)
    )

print("Dataset successfully split into training and validation sets!")
```

```python
import cv2
import os

# Paths (Google Drive)
original_videos_path = "/content/drive/MyDrive/KaggleDatasets/DFD_original sequences"
manipulated_videos_path = "/content/drive/MyDrive/KaggleDatasets/DFD_manipulated_sequences (1)"

# Save extracted frames directly in Google Drive
output_original = "/content/drive/MyDrive/KaggleDatasets/extracted_frames_original"
output_manipulated = "/content/drive/MyDrive/KaggleDatasets/extracted_frames_manipulated"

# Ensure output directories exist
os.makedirs(output_original, exist_ok=True)
os.makedirs(output_manipulated, exist_ok=True)

# Function to extract frames
def extract_frames(video_path, output_folder, max_frames=3000):
    total_frames = 0
    for video_file in os.listdir(video_path):
        video_file_path = os.path.join(video_path, video_file)

        if not video_file_path.endswith(('.mp4', '.avi', '.mov', '.mkv')):
            continue  # Skip non-video files

        cap = cv2.VideoCapture(video_file_path)
        frame_count = 0

        while cap.isOpened() and total_frames < max_frames:
            ret, frame = cap.read()
            if not ret:
                break

            frame_filename = f"{video_file}_frame_{frame_count:04d}.jpg"
            frame_filepath = os.path.join(output_folder, frame_filename)
            cv2.imwrite(frame_filepath, frame)

            frame_count += 1
            total_frames += 1

        cap.release()

        if total_frames >= max_frames:
            break  # Stop when 3000 frames are extracted

    print(f"Extracted {total_frames} frames to {output_folder}")

# Extract frames and save in Google Drive
extract_frames(original_videos_path, output_original, max_frames=3000)
extract_frames(manipulated_videos_path, output_manipulated, max_frames=3000)
```

```python
import cv2
import os
from tqdm import tqdm  # For progress bar

# Define paths
original_images_path = "/content/drive/MyDrive/KaggleDatasets/extracted_frames_original"
manipulated_images_path = "/content/drive/MyDrive/KaggleDatasets/extracted_frames_manipulated"

# Output folders (overwrite original images)
processed_original_path = "/content/drive/MyDrive/KaggleDatasets/processed_frames_original"
processed_manipulated_path = "/content/drive/MyDrive/KaggleDatasets/processed_frames_manipulated"

# Create processed folders if not exist
os.makedirs(processed_original_path, exist_ok=True)
os.makedirs(processed_manipulated_path, exist_ok=True)

# Function to preprocess images
def preprocess_images(input_folder, output_folder, img_size=(224, 224)):
    for img_file in tqdm(os.listdir(input_folder), desc=f"Processing {input_folder}"):
        img_path = os.path.join(input_folder, img_file)
        output_path = os.path.join(output_folder, img_file)

        try:
            img = cv2.imread(img_path)
            if img is None:
                print(f"Skipping corrupted file: {img_file}")
                continue

            img = cv2.resize(img, img_size)  # Resize to 224x224
            img = img / 255.0  # Normalize pixel values

            # Save the preprocessed image (overwrite original file)
            cv2.imwrite(output_path, img * 255)  # Convert back to 0-255 range for saving
        except Exception as e:
            print(f"Error processing {img_file}: {e}")

# Run preprocessing
preprocess_images(original_images_path, processed_original_path)
preprocess_images(manipulated_images_path, processed_manipulated_path)

print("✅ Preprocessing Complete! Images are resized and normalized.")
```
